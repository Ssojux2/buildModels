#### 표본 공간 (A Sample space)
* A Sample space is the set of all possible outcomes of an experiment. 
* 시행에서 발생 가능한 모든 경우의 집합

#### 사건 (An event)
* An event is a subset of the sample space.
* 표본 공간의 부분 집합
* 사건을 부분집합으로 생각하는 것에서 시작

#### 확률의 naive한 정의 (Naive definition of probability)

* 발생 가능한 모든 경우의 수(possible outcomes) : S와 크기가 같다.

#### 동전 두번 던지기 (flip a coin twice)

$$ 
    \begin{bmatrix}
    HH \\
    HT \\
    TH \\
    TT \\

$$

* TT 가 나오는 경우는 전체 4번 중 1번 이므로 $ \frac{1}{4} $ 임을 확인할 수 있다.

#### 굉장히 엄격한 규율이 적용되어있다.
* 모든 사건이 발생할 경우가 같은 확률로 나온다
* All outcomes are equally likely => 굉장히 엄격한 가정이다.
* 유한한 표본 공간 (A finite sample space)
* 항상 만족하는 것이 아니기 때문에 적용 불가능한 경우도 있다.

#### 셈 원리(Counting Principle) - 곱의 법칙 (Multiplication Rule)
* 발생 가능한 경우의 수가 $ n_1, n_2, ... , n_r $ 가지인 1, 2, ... , r 번에 시행에서 발생 가능한 모든 경우의 수는 $ n_1 \times n_2 \times ... \times n_r $ 이다.
* simple tree로 표현 할 수도 있다.


#### 이항계수 (Binominal Coefficent)
* Subsets of size K of group of n people (순서 상관 X)

$$ 
    \begin{pmatrix}
    n \\
    k \\
    \end{pmatrix}
    = \frac{n!}{(n-k)!k!}
$$

### Story proof - proof by interpretation 해석의 의한 증명
* 상황 해석을 통한 증명 (대수적 방법 보다 쉬울대가 있다)

* ex1) 
$ 
    \begin{pmatrix}
    n \\
    k \\
    \end{pmatrix}
    = 
    \begin{pmatrix}
    n \\
    n-k \\
    \end{pmatrix}
$

* ex2) $  n \times 
    \begin{pmatrix}
    n-1 \\
    k-1 \\
    \end{pmatrix}
    = k \times
    \begin{pmatrix}
    n \\
    k \\
    \end{pmatrix}
$


    * n 명중에 k명 뽑기, k 명 중에서 한 명을 회장으로 뽑는 문제로 해석
    * 다른 방법 같은 결과
    * 회장 먼저 뽑고 나머지 k-1 명 뽑기 // k 명을 뽑고 회장 뽑기


* ex3) $
    \begin{pmatrix}
    m+n \\
    k \\
    \end{pmatrix}
    = \sum_{j=0}^k
    \begin{pmatrix}
    m \\
    j \\
    \end{pmatrix}
    \begin{pmatrix}
    n \\
    k-j \\
    \end{pmatrix}
$
