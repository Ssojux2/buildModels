```
import tensorflow as tf
```

```
#shape of inputs in cnn : [batch_size, height, width, channels]
#shape of inputs in rnn : [batch_size, sequence_lenght, input_dim]

batch_size = 4
seq_len = 5
input_dim = 3

rnn_inputs = tf.random_normal([batch_size, seq_len, input_dim]) 
```

```
rnn_cell = tf.keras.layers.SimpleRNNCell(units=4)
gru_cell = tf.keras.layers.GRUCell(units=4)
lstm_cell = tf.keras.layers.LSTMCell(units=4)
```

```
outputs, state = tf.nn.dynamic_rnn(lstm_cell, rnn_inputs, dtype=tf.float32)
```
