```
import torch
```

```
x = torch.rand(441000, 1)
y = torch.rand(441000, 1)

print(x)
print(y)
```

```
z = torch.cat((x, y), 1)
print(z)
print(z.shape)
```

```
z1 = torch.stack((x, y))
print(z1)
print(z1.shape)
```

```
a = torch.rand(441000)
b = torch.rand(441000)

print(a.shape, b.shape)
```

```
c = torch.cat((a,b), 0)
print(c, c.shape)
```

```
a = a.reshape(441000, 1)
b = b.reshape(441000, 1)

print(a.shape, b.shape)
```

```
c = torch.cat((a,b), 1)
print(c, c.shape)
```
